```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from dftpy.td.utils import calc_spectra_mu
from dftpy.constants import Units
```

```python
dipole   = np.load('./dipole/dipole_24000.npy')
```

```python
time = dipole[:,0]
dipole_x = dipole[:,1]
dipole_y = dipole[:,2]
dipole_z = dipole[:,3]
```

```python
interval = time[1] - time[0]
direction = 0 
k = 1.0e-3
```

```python
plt.plot(time, dipole_x)
plt.xlabel('Time (au)')
plt.ylabel('Dipole Moment (au)')
```

```python
omega, spectra = calc_spectra_mu(dipole_x, interval, kick=k, sigma=0.002)
plt.plot(omega*Units.Ha, spectra, '-')
plt.xlabel('Frequency (eV)')
plt.ylabel('Intensity (au)')
plt.xlim(0, 11)
```
